```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

# As suggested by the error message, we can fetch the dataset from the original source:
data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep=r"\s+", skiprows=22, header=None)
data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
target = raw_df.values[1::2, 2]

# Now 'data' contains the features and 'target' contains the target variable.
# You can convert them to a pandas DataFrame if needed:
# boston_df = pd.DataFrame(data, columns=[...]) # Add appropriate column names
# boston_df['target'] = target
```

```python

# Column names for the Boston dataset
feature_names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT']

# Create the DataFrame
boston = pd.DataFrame(data, columns=feature_names)
boston['target'] = target

# Display the first few rows
display(boston.head())
```

**Notes**

Data Set Characteristics:

    :Attribute Information (in order):
        - CRIM     per capita crime rate by town
        - ZN       proportion of residential land zoned for lots over 25,000 sq.ft.
        - INDUS    proportion of non-retail business acres per town
        - CHAS     Charles River dummy variable (= 1 if tract bounds river; 0 otherwise)
        - NOX      nitric oxides concentration (parts per 10 million)
        - RM       average number of rooms per dwelling
        - AGE      proportion of owner-occupied units built prior to 1940
        - DIS      weighted distances to five Boston employment centres
        - RAD      index of accessibility to radial highways
        - TAX      full-value property-tax rate per $10,000
        - PTRATIO  pupil-teacher ratio by town
        - B        1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town
        - LSTAT    % lower status of the population
        - MEDV     Median value of owner-occupied homes in $1000s

    :Missing Attribute Values: None.

```python
## Independent and Dependent Features

X = boston.iloc[:, :-1]
y = boston.iloc[:, -1]
```

```python
X.head()
```

```python
# Train Test and Split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=47)
```

```python
X_train
```

```python
y_train
```

```python
## Stadardize the dataset IMPORTANT.
'''
# Internally we use Gradient Descent and our main aim is to come at Global Minima, So to make sure all our
Independant features unit should be on a same scale because of that the conversions will happend quicky.
'''

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
import pickle 
pickle.dump(scaler, open('scaling.plk', 'wb'))
```

```python
X_train
```

```python
X_test
```

## **Model Training**

```python
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
```

```python
regression.fit(X_train, y_train)
```

```python
## Print the Cofficient and Intercept
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
## On which parameters the model has been trained
regression.get_params()
```

 Prediction with Test

```python
reg_pred = regression.predict(X_test)
```

```python
reg_pred
```

```python
## Plot a scatter plot for the prediction
plt.scatter(y_test, reg_pred)
```

```python
## Residuals "ERROR"
residuals = y_test - reg_pred
residuals
```

### **Assumptions**

```python
## Ploting the residuals

sns.displot(residuals, kind="kde")
```

```python
## Scater plot with respect to prediction and residuals
## Uniform Distribution
plt.scatter(reg_pred, residuals)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print(mean_absolute_error(y_test, reg_pred))
print(mean_squared_error(y_test, reg_pred))
print(np.sqrt(mean_squared_error(y_test, reg_pred)))
```

### **R square and adjusted R square**
R^2 = 1 - SSR/SST (SSR=Sum of squares of residuals, SST=total sum of squares)

```python
from sklearn.metrics import r2_score
score = r2_score(y_test, reg_pred)
print(score)
```

Adjusted R2 = 1-[(1-R2)*(n-1)/(n-k-1)]

```python
# Display
1 - (1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```
